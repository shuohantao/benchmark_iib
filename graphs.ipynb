```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
ga = pickle.load(open("gano_mmds_4904578.pkl", "rb"))
resn = pickle.load(open("res_sno_mmds_64360.pkl", "rb"))
ca = pickle.load(open("car_mmds.pkl", "rb"))
```

```python
res = [2*i for i in range(14, 28)]+[56]
plt.plot(res, ga, label="GA")
plt.plot(res, resn,label="RES")
plt.plot(res, ca, label="CA")
plt.xlabel("Resolutions")
plt.ylabel("MMD")
# plt.yscale("log")
plt.legend()
plt.show()
```
