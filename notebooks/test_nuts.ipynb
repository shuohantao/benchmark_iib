```python
from nuts import nuts6
import numpy as np
```

```python
gamma = 0.1
```

```python
def annealed_normal(theta):
    A = np.eye(len(theta))
    grad = - np.dot(theta, A) * gamma
    logp = 0.5 * gamma * np.dot(grad, theta.T)
    return logp, grad
```

```python
D = 2
M = 10000
Madapt = 5000
theta0 = np.random.normal(0, 1, D)
delta = 0.2
```

```python
samples, lnprob, epsilon = nuts6(annealed_normal, M, Madapt, theta0, delta, progress=True)
```

```python
import matplotlib.pyplot as plt
plt.plot(samples[:, 0], samples[:, 1], 'r+')
plt.show()
```
